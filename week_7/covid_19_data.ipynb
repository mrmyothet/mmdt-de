```python
import pandas as pd
import requests 
```

```python
url = 'https://raw.githubusercontent.com/owid/covid-19-data/refs/heads/master/public/data/latest/owid-covid-latest.json'
response = requests.get(url)
data = response.json()

# result_df = pd.json_normalize(data)
df = pd.DataFrame(data)
covid_df = df.transpose()
covid_df
```

```python
covid_df = pd.DataFrame()

# row = data["AFG"]
# row_df = pd.json_normalize(row)
# type(row_df)

# row_df.insert(0, "country_code", "AFG")
# row_df


for key in data.keys():
    row = data[key]
    row_df = pd.json_normalize(row)
    # row_df['country_code'] = key
    row_df.insert(0, "country_code", key)
    covid_df = pd.concat([covid_df, row_df], ignore_index=True)

covid_df
```
